### 영어 Word2Vec 만들기

- 데이터 : ted_en-20160408.xml 

전처리 진행하기


```python
import gensim
import re
from lxml import etree
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
#xml 파일 읽어오기
targetXML = open('ted_en-20160408.xml','r',encoding = 'utf-8')


target_text = etree.parse(targetXML)


#XML 파일로부터 <content>와 </content> 사이의 내용만 가져옴
parse_text = '\n'.join(target_text.xpath('//content/text()'))
```

```python
parse_text
```

```python
#content 중간에 등장하는 (Audio),  (Laughter) 등의 배경음 부분 제거
content_text = re.sub(r'\([^)]*\)','',parse_text)
content_text
```

```python
#입력 코퍼스에 대해서 NLTK를 이용하여 문장 토큰화 수행
sent_text = sent_tokenize(content_text)
```

```python
normalized_text = []

#각 문장에 대하여 구두점을 제거하고,,, 대문자를 소문자로 변환
for string in sent_text : 
    tokens = re.sub(r"[^a-z0-9]+"," ",string.lower())
    normalized_text.append(tokens)
    
normalized_text 
```

```python
result =[]
result = [word_tokenize(sentence) for sentence in normalized_text] 
#각 문장에 대해서 NLTK를 이용하여 단어 토큰화를 수행

print(result[0])
```

```python
from gensim.models.word2vec  import Word2Vec

model =Word2Vec(result,vector_size=100, window=10,min_count=10,workers=-1,sg=0)
model.wv.similarity("man","woman")
```

```python
model.wv.most_similar("man")
```

```python
#불용어 제거하면서 토큰화
from nltk.corpus import stopwords 

data_cln= []

stop_words = set(stopwords.words('english'))


for i in normalized_text:
    sentence = word_tokenize(i)
    result = []
    
    for word in sentence:
        if word not in stop_words : 
            if len(word) >2 :
                result.append(word)
                
                
    data_cln.append(result)
    
print(data_cln[0])
```

```python
from gensim.models.word2vec  import Word2Vec

model2 =Word2Vec(data_cln,vector_size=100, window=10,min_count=10,workers=-1,sg=0)
model2.wv.similarity("man","woman")
```
