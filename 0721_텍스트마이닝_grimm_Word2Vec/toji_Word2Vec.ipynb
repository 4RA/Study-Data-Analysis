```python
import codecs
from bs4 import BeautifulSoup
from konlpy.tag import Okt
from gensim.models.word2vec import Word2Vec
```

```python
#utf-16 인코딩으로 파일을 열고 글자 출력 
fp = codecs.open("BEXX0003.txt",encoding="utf-16")
soup = BeautifulSoup(fp,"html.parser")
body = soup.select_one("body > text")
text = body.getText()
```

```python
#텍스트를 한 줄씩 처리

okt = Okt()
results = []
lines = text.split("\n")

for line in lines :
    #형태소 분석하기 
    #단어의 기본형 사용
    
    malist = okt.pos(line,norm=True,stem = True)
    r  = []
    
    for word in malist:
        #어미/조사/구두점 등은 대상에서 제외
        if not word[1] in ["Josa","Eomi","Punctuation"]: #word[1] 은 형태소를 뜻함.
            r.append(word[0])
            
    results.append(r)
    
print(results)
```

```python
#텍스트를 한 줄씩 처리

okt = Okt()
results = []
lines = text.split("\n")

for line in lines :
    #형태소 분석하기 
    #단어의 기본형 사용
    
    malist = okt.pos(line,norm=True,stem = True)
    r  = []
    
    for word in malist:
        #어미/조사/구두점 등은 대상에서 제외
        if not word[1] in ["Josa","Eomi","Punctuation"]: #word[1] 은 형태소를 뜻함.
            #명사만 가져오겠다 하면 if word[1] == "None"
            r.append(word[0])
            
    results.append(r)
    
print(results)
```

```python
#Word2Vec 모델 생성
model = Word2Vec(results,vector_size = 200,window =10 ,min_count=2,sg = 1)

model.wv.most_similar("인사")
```
