### <연습문제>
- #### Age의 NaN 값을 다음과 같이 처리하시오.
    - 생존자(Survived = 1)는 생존자 나이의 평균으로 대체
    - 사망자(Survived = 0)는 사망자 나이의 평균으로 대체

```python
survive_1 = train[train["Survived"]==1]['Age'].mean()
survive_0 = train[train["Survived"]==0]["Age"].mean()

print(survive_1,survive_0)

train[train['Survived']==1]['Age'].fillna(survive_1)
train[train['Survived']==0]['Age'].fillna(survive_0)
```
