## (추가설명)robots.txt

###### robots.txt란?

- 개발한 웹 페이지에서 로봇들이 크롤링을 할 수 있는 접근 권한을 제한하는 문서
- 물리적으로 제한하지는 않지만 권고하는 것으로 로봇이 파일을 읽고 접근하는 것을 중지하길 바라는 것에서 설정함
- 접근하지 못하도록 써놓더라도 다른사람들이 접근할 수는 있음
- 위치는 항상 사이트의 루트 디렉토리에 위치한다. ex) www.naver.com/robots.txt

>User-agent: *
>
>Allow: / 

모든 로봇들이 모든 페이지에 접근이 가능한 것

>User-agent: *
>
>Disallow: /

모든 로봇들이 접근하지 못하게 하는 것

> User-agent: MaderiBot
>
> Disallow: /

Maderi bot은 접근할 수 없다. (특정 로봇만 거부)

> User-agent: *             
>    
> Disallow: /board/        
    
모든 로봇에 대해서 board 디렉토리의 접근을 거부한다    

## (추가설명2) 서버에 부하주지 않기

크롤링으로 서버에 부하를 주는 것은 네트워크 요청을 많이 준다는 것인데
요청이 너무 많아서 서버가 견딜 수 없다면 다른 요청들을 처리하기 힘들거나
심지어는 서버가 다운되서 서비스를 이용할 수 없게 될지도 모른다.

유명한 게임을 해본사람들은 알겠지만 대규모 이벤트로 사람들이 몰리면 서버가
제대로 작동하지 못하고 게임이 느려지거나 심지어는 로그아웃되는 경우도 있을 것을
경험한 적이 있을 것이다.

네트워크를 요청하는것도 같은 방식으로 요청 받는 사이트에 문제가 생길 수 있다

일반적으로 서비스에서 크롤링을 바라는 경우는 없다.
크롤링은 일반적으로 대량으로 데이터를 가져가기 위해서 수많은 요청들을 짧은 시간동안 하게 된다.

반복문을 통해서 요청을 계속하게되면 서비스에 문제가 생길 수 있기 때문에
가급적 요청마다 일정 간격을 주는 것이 필요하기도 하다.

방법은 실제로 크롤링을 할 때 설명하도록 한다.

### (추가설명3) 크롤링 데이터 사용 유의
크롤링을 하는 이유는 모두가 다르다.
마케팅 성과를 측정하기 위해서,
새로운 제품을 기획하기 위해서 등 모두가 데이터를 사용하는 목적이 다르다.

하지만 크롤링을 통해 데이터를 수집하는 것은 조금은 조심해야할 필요가 있다.

종종 경쟁사의 사이트의 데이터를 크롤링하다가 소송까지 가는 경우들도 있고,
특정 api에서 정책상 데이터를 직접, 재가공판매해서는 안 된다라고 명시해두기도 한다.

물론 크롤링이 법적인 문제가 된다고 명확하게 말할 수는 없지만 특정 경우에는 문제가
될 수 있기때문에 위에 추가설명에서 말한 2가지는 확인해서 최대한 문제가 되지 않도록
크롤링하는 것이 좋다.
